```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_actual = pd.read_csv('actual.csv')
df_ARIMA = pd.read_csv('ARIMA.csv')
df_omp = pd.read_csv('omp_cds_dt.csv')
df_blended = pd.read_csv('blended.csv')
```

```python
print(df_ARIMA)
```

```python
fig, ax = plt.subplots()
ax.plot(df_actual['target1d'], 'k', label='Actual', linewidth=4)
ax.plot(df_ARIMA['y_pred'], label='Auto ARIMA')
ax.plot(df_omp['y_pred'], label='omp_cds_dt')
ax.plot(df_blended['y_pred'], label='Blended')
ax.legend()
plt.title('AAPL Return Prediction Performance of PyCaret Models')
dates = ['4-18-22', '4-19-22', '4-20-22', '4-21-22', '4-22-22', '4-25-22', '4-26-22']
plt.xticks([0, 1, 2, 3, 4, 5, 6], dates, rotation='vertical')
plt.xlabel('Date')
plt.ylabel('Return')

plt.show()
```
